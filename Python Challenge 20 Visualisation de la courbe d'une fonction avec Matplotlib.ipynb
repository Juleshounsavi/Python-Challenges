```python
#Impotation des biblioth√®ques
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#definition de la fonction
def f(x):
    return (np.log(x))/(x**2 + 1)

#Declaration des variables
var = np.arange(0,250,1)
x = var[var!=0]
y = f(x)

#Graphique de la fonction
plt.plot(x,y)
plt.axvline(x=0, color='r', linestyle='--', label='Asymptote verticale (x=0)')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
