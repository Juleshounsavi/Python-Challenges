```python
#Plots of some math functions using python libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import cos, sin, tan, log, exp, sqrt, pi
```

# 1-Sin(x) function

```python
x_trigo = np.linspace(-2*pi, 2*pi, 1000)
y_sin = []
for i in x_trigo:
    y_sin.append(sin(i))
```

```python
plt.plot(x_trigo, y_sin)
plt.title("Sin(x) function representation")
plt.xlabel("X")
plt.ylabel("Sin(X)")
plt.show()
```

# 2-Cos(x) function

```python
x_trigo = np.linspace(-2*pi, 2*pi, 1000)
y_cos = []
for i in x_trigo:
    y_cos.append(cos(i))
```

```python
plt.plot(x_trigo, y_cos)
plt.title("Cos(x) function representation")
plt.xlabel("X")
plt.ylabel("Cos(X)")
plt.show()
```

# 3-Tan(x) function

```python
x_trigo = np.linspace(-2*pi, 2*pi, 1000)
y_tan = []
for i in x_trigo:
    y_tan.append(tan(i))
```

```python
plt.plot(x_trigo, y_tan)
plt.title("Tan(x) function representation")
plt.xlabel("X")
plt.ylabel("Tan(X)")
plt.ylim(-10, 10)
for i in range(-2, 2):
    plt.axvline(x=(i + 0.5) * np.pi, color='r', linestyle='--', linewidth=2)
plt.show()
```
